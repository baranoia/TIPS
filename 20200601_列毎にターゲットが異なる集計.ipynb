直近6か月ターゲット月のカラム名は hist201910 ～ hist202003。  
これを求めて、カラムに入れておきたい

```python
df["APP_bef0"] = np.nan
df["APP_bef1"] = np.nan
df["APP_bef2"] = np.nan
df["APP_bef3"] = np.nan
df["APP_bef4"] = np.nan
df["APP_bef5"] = np.nan
df["APP_bef6"] = np.nan

for i in df.index:
    year  = df[df["ID"] == i]["APPDATE"].dt.year
    month = df[df["ID"] == i]["APPDATE"].dt.month
    # df["tmp"] = datetime.datetime(year, month, 1) + relativedelta(months = -0)

    APP_bef0 = datetime.datetime(year, month, 1) + relativedelta(months = -0)
    APP_bef1 = datetime.datetime(year, month, 1) + relativedelta(months = -1)
    APP_bef2 = datetime.datetime(year, month, 1) + relativedelta(months = -2)
    APP_bef3 = datetime.datetime(year, month, 1) + relativedelta(months = -3)
    APP_bef4 = datetime.datetime(year, month, 1) + relativedelta(months = -4)
    APP_bef5 = datetime.datetime(year, month, 1) + relativedelta(months = -5)
    APP_bef6 = datetime.datetime(year, month, 1) + relativedelta(months = -6)
    
    #df.loc[i, "APP_bef0"] = datetime.datetime(year, month, 1) + relativedelta(months = -0)
    #df.loc[i, "APP_bef1"] = datetime.datetime(year, month, 1) + relativedelta(months = -1)
    #df.loc[i, "APP_bef2"] = datetime.datetime(year, month, 1) + relativedelta(months = -2)
    #df.loc[i, "APP_bef3"] = datetime.datetime(year, month, 1) + relativedelta(months = -3)
    #df.loc[i, "APP_bef4"] = datetime.datetime(year, month, 1) + relativedelta(months = -4)
    #df.loc[i, "APP_bef5"] = datetime.datetime(year, month, 1) + relativedelta(months = -5)
    #df.loc[i, "APP_bef6"] = datetime.datetime(year, month, 1) + relativedelta(months = -6)
    
    df.loc[i, "APP_bef0"] = APP_bef0.strftime("hist%Y%m")
    df.loc[i, "APP_bef1"] = APP_bef1.strftime("hist%Y%m")
    df.loc[i, "APP_bef2"] = APP_bef2.strftime("hist%Y%m")
    df.loc[i, "APP_bef3"] = APP_bef3.strftime("hist%Y%m")
    df.loc[i, "APP_bef4"] = APP_bef4.strftime("hist%Y%m")
    df.loc[i, "APP_bef5"] = APP_bef5.strftime("hist%Y%m")
    df.loc[i, "APP_bef6"] = APP_bef6.strftime("hist%Y%m")
```
