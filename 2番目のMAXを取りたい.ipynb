```python
# ノック15 金額欠損の補完をしよう
```

```python
# item_price にnullがある
uriage_data.isnull().any(axis=0)
```

```python
uriage_data["item_price"].isnull().sum()
```

```python
# 同じ商品は常に同一価格なので
# 同じ商品から補完する
# テキストとやり方はちがうけどいいや
uriage_data[["item_name", "item_price"]].sort_values(by="item_name")
```

```python
price_map ={} 
for name in uriage_data["item_name"].sort_values().unique():
    for price in list(uriage_data[uriage_data["item_name"] == name]["item_price"].value_counts().index):
        print(name,price)
        price_map[name] = price
```

```python
def retPrice(key):
    return price_map[key]
```

```python
# チェックしたいので別カラム名で入れて
uriage_data["item_price_"] = uriage_data["item_name"].apply(retPrice)
```

```python
# チェック
# 埋めたのと元の値で一致しないのは、元の値がnullのレコードのみ
#uriage_data[ uriage_data["item_price"] != uriage_data["item_price_"] ][["item_name", "item_name", "item_price", "item_price_"]]
uriage_data[ uriage_data["item_price"] != uriage_data["item_price_"] ]["item_price"].value_counts(dropna=False)
```

```python
uriage_data["item_price"] =  uriage_data["item_price_"]
```

```python
uriage_data["item_price"].isnull().sum()
```

```python
uriage_data = uriage_data.drop(["item_price_"], axis=1,)
```

```python
# さいごにチェック
for name in uriage_data["item_name"].sort_values().unique():
    pricemin = uriage_data[uriage_data["item_name"] == name] ["item_price"].min()
    pricemax = uriage_data[uriage_data["item_name"] == name] ["item_price"].max()
    print(name, "min=", pricemin,  "max=", pricemax, "一致する？",  pricemin==pricemax)
```

```python
pd.set_option("display.max_row", 5000)
kokyaku_data["登録日"]
```

```python
# 42782など日付でないものがある
# 原因はもともとのcsvの書式が混在しているため。
```

```python
# isdigit()で数字かどうか判断して、色付け
# 42782みたいになってるのは22個
flg_is_serial = kokyaku_data["登録日"].astype("str").str.isdigit()
display(flg_is_serial.head(5))
display(flg_is_serial.sum())
```

```python
#kokyaku_data.loc[flg_is_serial, "登録日"]
#kokyaku_data.loc[flg_is_serial, "登録日"].astype(float)
#pd.to_datetime(kokyaku_data.loc[flg_is_serial, "登録日"].astype(float), unit="D")
fromSerial = pd.to_datetime(kokyaku_data.loc[flg_is_serial, "登録日"].astype(float), unit="D")
fromSerial
```

```python
# 日付になっているレコードも書式統一（datetime）のために処理する
fromString = pd.to_datetime(kokyaku_data.loc[~flg_is_serial, "登録日"])
fromString
```

```python
# 直したレコードを結合
display(fromSerial.head(5))
display(fromString.head(5))
pd.concat([fromSerial, fromString]).head(5)
```

```python
kokyaku_data["登録日_"] = pd.concat([fromSerial, fromString])
kokyaku_data[["登録日","登録日_"]].head(10)
```

```python
kokyaku_data["登録日"] = kokyaku_data["登録日_"]
kokyaku_data = kokyaku_data.drop("登録日_", axis=1)
```
