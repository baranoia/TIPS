### HashingEncoder
category_encoders.HashingEncoder[▼DOC](
http://contrib.scikit-learn.org/category_encoders/hashing.html
) 
>次元/精度を構成可能な多変量ハッシュの実装。  
このエンコーダの利点は、監視されたカテゴリの辞書を維持しないことです。  
その結果、エンコーダーはサイズが大きくならず、  
データスコアリング中に新しい値を受け取ることができます。  
max_processとmax_sampleを手動で設定する前に、  
それらがどのように機能するかを読むことが重要です。  
マルチプロセッシングが問題を引き起こす可能性があるため、  
Windowsでは「max_process」のデフォルト値は1です。次を 参照してください   https : //github.com/scikit-learn-contrib/categorical-encoding/issues/215 https://docs.python.org/ 2 / library / multiprocessing.html？highlight = process＃windows

```python
df = pd.DataFrame({
        "COLOR" : ["Red", "Yellow", "Red", "Red", "Blue"],
        "SIZE" : ["S", "M", "L", "S", "S"],
        "WEIGHT" : [200, 100, 300, 300, 300],
        "HARDNESS" : ["hard", "soft", "medium", "hard", "soft"],
        "OUTCOME" : [1,2,0,0,0] })
df
```

```python
# 説明変数X と 目的変数Yに分ける
df_X = df[["COLOR", "SIZE", "WEIGHT","HARDNESS"]].copy()
df_Y = df[["OUTCOME"]].copy()
```

```python
df_X_test = pd.DataFrame({
        "COLOR" : ["Red", "Yellow", "Red", "Green"],
        "SIZE" : ["S", "M", "L", "XL", ],
        "WEIGHT" : [200, 100, 300, 400],
        "HARDNESS" : ["hard", "soft", "medium", "maxhard"] })
df_X_test
```

・Encodeしてみる

```python
print("エンコード前")
display(df_X)

print("エンコード後")
enc = HashingEncoder()
enc.fit_transform( df_X )
```
