```python
import pandas as pd
import numpy as nm
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
df1 = pd.DataFrame(boston.data, columns=boston.feature_names)
df2 = pd.DataFrame(boston.target, columns=["Y"])
```

```python
df1
```

```python
df2
```

```python
df = df1.join(df2)
df
```

```python
# where のテストができればいいだけなので、先頭10行だけ切り取る
df = df.loc[0:9,:]
dfbk = df.copy()
```

```python
df
```

### ◆DataFrameに対するwhere

```python
# where
# 条件に合致しないものをNan表示する
df["INDUS"].where(df["INDUS"] > 7.5)
```

```python
# where
# 条件に合致しないものを第2引数に置換
df["INDUS"].where(df["INDUS"] > 7.5, 999)
```

```python
# where
# 条件に合致しないものを第2引数に置換
df["INDUS"] = df["INDUS"].where(df["INDUS"] > 7.5, 999)
df
```

### DataFrameに対するmask

```python
# mask
# 条件に合致するものをNan表示
df["ZN"].mask(df["ZN"] < 12)
```

```python
# mask
# 条件に合致するものを第２引数に置換
df["ZN"].mask(df["ZN"] < 12, 1)
```

```python
# mask
# 条件に合致するものを第２引数に置換
df["ZN"] = df["ZN"].mask(df["ZN"] < 12, 1)
df
```
