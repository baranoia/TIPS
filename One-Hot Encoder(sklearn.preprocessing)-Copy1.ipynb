```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({
        "COLOR" : ["Red", "Yellow", "Red", "Red", "Blue"],
        "SIZE" : ["S", "M", "L", "S", "S"],
        "WEIGHT" : [200, 100, 300, 300, 300],
        "HARDNESS" : ["hard", "soft", "medium", "hard", "soft"],
        "OUTCOME" : [1,2,0,0,0] })
df
```

```python
# 説明変数X と 目的変数Yに分ける
df_X = df[["COLOR", "SIZE", "WEIGHT","HARDNESS"]].copy()
df_Y = df[["OUTCOME"]].copy()
```

```python
df_X_test = pd.DataFrame({
        "COLOR" : ["Red", "Yellow", "Red", "Green"],
        "SIZE" : ["S", "M", "L", "XL", ],
        "WEIGHT" : [200, 100, 300, 400],
        "HARDNESS" : ["hard", "soft", "medium", "maxhard"] })
df_X_test
```

- エンコードしてみる

```python
print("エンコード前")
display(df_X[["COLOR", "SIZE", "HARDNESS"]])

print("エンコード後")

# sparse=False でndarrayのかたちで返るようにする
oh = OneHotEncoder(sparse=False)
pd.DataFrame( oh.fit_transform( df_X[["COLOR", "SIZE", "HARDNESS"]] ) )
# どのカラムがどのカラムに対応しているかぱっと見わからない
```

- get_dummiesなら、どのカラムがどのカラムと対応しているのかわかりやすい

```python
print("エンコード前")
display(df_X[["COLOR", "SIZE", "HARDNESS"]])

print("エンコード後")
pd.get_dummies( df_X[["COLOR", "SIZE", "HARDNESS"]] )
```

- でもOneHotEncoderなら、fit＆transformができる  
 パラメタを指定することでfit時にはなかった値を無視できる

```python
print("エンコード前")
display(df_X_test[["COLOR"]])

print("エンコード後")
# handle_unknown="ignore"でfit時にはなかった値を無視
oh = OneHotEncoder(sparse=False, handle_unknown="ignore")
oh.fit(df_X[["COLOR"]])
pd.DataFrame( oh.transform( df_X_test[["COLOR"]] ) )
```

- nanがある場合のgetDummies  
　nanが無視されるのでnanも1つの列としたい場合はnanを文字列"null"に置換するなどが必要

```python
df_X_tmp = pd.DataFrame({
        "COLOR" : ["Red", "Yellow", "Red", np.nan, "Blue"],
        "HARDNESS" : ["hard", np.nan, "medium", "hard", "soft"] })

print("エンコード前")
display(df_X_tmp)

print("エンコード後")
pd.get_dummies(df_X_tmp)
```

- nanがある場合のOneHotEncoder  
　エラーになる

```python
df_X_tmp = pd.DataFrame({
        "COLOR" : ["Red", "Yellow", "Red", np.nan, "Blue"]})

#print("エンコード前")
#display(df_X_tmp)

try:
    print("エンコード後")
    oh = OneHotEncoder(sparse=False)
    pd.DataFrame( oh.fit_transform( df_X_tmp[["COLOR"]] ) )
except ValueError as e:
    print("エラー:", e)
```
